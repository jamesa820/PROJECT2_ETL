```python
sales_revenue=pd.read_csv('Marijuana_Sales_Revenue_in_Colorado.csv')
sales_revenue.head()
```

```python
simple_sales = sales_revenue.drop(columns=['Med_Blank_Code', 'Rec_Blank_Code'])
simple_sales.head()
```

```python
simple_sales.info()
```

```python
simple_sales.dropna(thresh=2)
simple_sales.head()
```

```python
clean_sales = simple_sales.dropna(how="all", subset=["Rec_Sales","Med_Sales"])
clean_sales.head()
```

```python
clean_sales.info()
```

```python
all_rec_sales = simple_sales.dropna(how="any", subset=["Rec_Sales"])
all_rec_sales.head()
```

```python
all_med_sales = simple_sales.dropna(how="any", subset=["Med_Sales"])
all_med_sales.head()
```
